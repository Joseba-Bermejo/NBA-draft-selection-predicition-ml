## Now that I have rookies classes between 2018-2022 with their PER I will merge them all together

```python
# Combine the DataFrames
rookie_per = pd.concat([merged_18_19, merged_19_20, merged_20_21, merged_21_22], ignore_index=True)
rookie_per
```

```python
# Now that I have this, I need to filter for rookies who played at least 40 games (half of the games in the season)

filt_rookie_per = rookie_per[rookie_per['G'] >= 40].reset_index(drop=True)
filt_rookie_per
```

```python
# Then I need to download college player stats for those years and keep all numerical columns that I deem valuable

df_college = pd.read_csv('CollegeBasketballPlayers2009-2021 2.csv')
df_college
```

```python
# Check columns names

df_college.columns.tolist()
```

```python
df_college.isnull().mean().sort_values(ascending=False).head(10)
```

```python
cols_to_keep = ['player_name', 'year', 'pick', 'pts', 'ast', 'treb', 'stl', 'blk', 'GP', 'usg', 
                'eFG', 'TS_per', 'FT_per', 'twoP_per','TP_per', 'blk_per','stl_per', 'ast/tov', 'mp']
df_col_clean = df_college[cols_to_keep]
```

```python
df_col_clean
```

```python
# Then I need to merge (inner) the college stats with nba rookie dataframe.

col_rookie_per = pd.merge(
    filt_rookie_per,
    df_col_clean,
    left_on='Player',
    right_on='player_name',
    how='inner'
)
```

```python
col_rookie_per
```

```python
# Now I want to keep the latest year if a player spend more than one year in college

college_rookie_per = col_rookie_per.loc[col_rookie_per.groupby('Player')['year'].idxmax()].reset_index(drop=True)
```

```python
college_rookie_per
```

```python
# Drop specific columns
columns_to_drop = ['player_name', 'blk_per', 'stl_per']
college_rookie_per = college_rookie_per.drop(columns=columns_to_drop, errors='ignore')
```

```python
college_rookie_per
```

```python
# Drop rows with null values in the 'pick' column, which mean the player wasn't selected for the NBA draft
col_rookie_per_cl = college_rookie_per.dropna(subset=['pick'])
```

```python
col_rookie_per_cl
```

```python
col_rookie_per_cl = col_rookie_per_cl.reset_index(drop=True)
```

```python
# Rename selected columns
col_rookie_per_cl = col_rookie_per_cl.rename(columns={
    'G': 'NBA_games',
    'GP': 'college_games',
    'stl': 'stlpg',
    'pts': 'ppg',
    'ast': 'apg',
    'treb': 'rpg',
    'blk': 'blkpg',
    'mp': 'min_played'
})
```

```python
col_rookie_per_cl.to_csv('cleaned_col_nba_rookie_data.csv', index=False)
```
